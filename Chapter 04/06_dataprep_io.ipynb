## read_json

Notice how the array from the first record get's split into multiple rows?

```python
import azureml.dataprep as dprep

# Read JSON files
dataflow = dprep.read_json(path='./data/*.json')
dataflow.head()
```

## read_parquet

```python
# Read single paqrquet files individually
dataflow = dprep.read_parquet_file(path='./data/*.parquet')
dataflow.head()
```

## read_parquet_dataset

Notice how the last column contains the directory information? This is called a paration column.

```python
# Read nested folders of parquet files
dataflow = dprep.read_parquet_dataset(path='./data/warehouse/telemetry')
dataflow.head()
```

# Output

```python
dataflow = dprep.auto_read_file(path='./data/train.csv')
```

```python
# Write data as Parquet file
dataflow.write_to_parquet(file_path='./outputs/train.parquet', single_file=True).run_local()
```

```python
# Write all partitions as parquet folder with _SUCCESS file
dataflow.write_to_parquet(directory_path='./outputs/parquet').run_local()
```
